```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("train.csv")
df.head()
```

```python
df.drop(['id', 'Name','CGPA'], axis=1, inplace=True)
```

```python
df['Total_Pressure'] = df[['Academic Pressure', 'Work Pressure']].sum(axis=1, skipna=True)
df['Overall_Satisfaction'] = df[['Study Satisfaction', 'Job Satisfaction']].fillna(0).sum(axis=1)
```

```python
df.head()
```

```python
df.drop(['Academic Pressure', 'Work Pressure', 'Study Satisfaction', 'Job Satisfaction'], axis=1, inplace=True)
```

```python
df.isna().sum()
```

```python
df['Profession'].fillna('Student', inplace=True)
```

```python
df.isna().sum()
```

```python
df.dropna(axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt

# 1. Total Pressure Index
print("🔵 Total Pressure Index (Mean):", df['Total_Pressure'].mean())
df['Total_Pressure'].plot(kind='hist', bins=12, color='darkcyan', edgecolor='black')
plt.title("Total Pressure Distribution")
plt.xlabel("Total Pressure Score")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# 2. Overall Satisfaction Index
print("🟡 Overall Satisfaction Index (Mean):", df['Overall_Satisfaction'].mean())
plt.scatter(df['Total_Pressure'], df['Overall_Satisfaction'], alpha=0.5, c='tomato')
plt.title("Satisfaction vs Pressure")
plt.xlabel("Total Pressure")
plt.ylabel("Overall Satisfaction")
plt.grid(True)
plt.show()
```

```python
# 3. Depression Rate
depression_rate = df['Depression'].mean()
print("🔴 Depression Rate:", round(depression_rate * 100, 2), "%")
df['Depression'].value_counts().plot(kind='bar', color=['skyblue','salmon'])
plt.title("Depression Distribution")
plt.xticks([0,1], ['No', 'Yes'], rotation=0)
plt.ylabel("Count")
plt.grid(axis='y')
plt.show()
```

```python
# 4. Suicidal Thoughts Prevalence
df['has_suicidal_thoughts'] = df['Have you ever had suicidal thoughts ?'].map({'Yes': 1, 'No': 0})
suicidal_rate = df['has_suicidal_thoughts'].mean()
print("🧨 Suicidal Thoughts Prevalence:", round(suicidal_rate * 100, 2), "%")
df['has_suicidal_thoughts'].value_counts().plot(kind='pie', labels=['No', 'Yes'], autopct='%1.1f%%', colors=['lightgreen','red'], startangle=90)
plt.title("Suicidal Thoughts Distribution")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report


# ✅ Feature selection
features = [
    'Age',
    'Work/Study Hours',
    'Total_Pressure',
    'Overall_Satisfaction',
    'Sleep Duration',
    'Gender',
    'City',
    'Working Professional or Student',
    'Profession',
    'Family History of Mental Illness',
    'has_suicidal_thoughts'
]

# 🔄 Label encode categorical columns
cat_cols = [
    'Sleep Duration',
    'Gender',
    'City',
    'Working Professional or Student',
    'Profession',
    'Family History of Mental Illness'
]

le = LabelEncoder()
for col in cat_cols:
    if col in df.columns:
        df[col] = le.fit_transform(df[col].astype(str))

# 🧪 Define X and y
X = df[features]
y = df['Depression']

# 🧩 Train-Test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,  random_state=45)
```

```python
model = xgb.XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    random_state=45,
    
    # 🌿 Tree structure
    max_depth=5,                # Controls tree depth to prevent overfitting
    min_child_weight=4,         # Minimum sum of instance weight needed in a child
    gamma=0.2,                  # Minimum loss reduction required to make a split

    # 🔄 Regularization
    subsample=0.5,              # Row sampling ratio
    colsample_bytree=0.7,       # Feature sampling ratio

    # 🚀 Learning
    learning_rate=0.1,          # Step size shrinkage
    n_estimators=200            # Number of boosting rounds
)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("✅ Accuracy Score:", accuracy_score(y_test, y_pred))
print("📋 Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# ⚡ Generate and plot the confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot(cmap='Blues', values_format='d')
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

# 🔮 Get predicted probabilities for positive class (Depression = 1)
y_proba = model.predict_proba(X_test)[:, 1]

# 🎯 Compute ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
roc_auc = roc_auc_score(y_test, y_proba)

# 📊 Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
